```python
# k-means
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


# 绘图函数
def show_cluster(data, cluster, centroids=None):
    # dataset：数据
    # centroids：聚类中心点的坐标
    # cluster：每个样本所属聚类
    # 不同种类的颜色，用以区分划分的数据的类别
    colors = np.array(['blue', 'red', 'green', 'purple'])
    # 画出所有样例
    plt.scatter(data[:, 0], data[:, 1], color=colors[cluster])

    # 画出中心点
    if centroids is not None:
        K = len(centroids)
        plt.scatter(centroids[:, 0], centroids[:, 1], color=colors[:K], marker='+', s=150)

    plt.show()

dataset = np.loadtxt('kmeans_data.csv', delimiter=',')
print('数据集大小', len(dataset))
show_cluster(dataset, np.zeros(len(dataset), dtype=int))   

kmeans = KMeans(n_clusters=4, random_state=0).fit(dataset)

show_cluster(dataset, kmeans.labels_, kmeans.cluster_centers_)

```
