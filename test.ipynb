```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler

# 假设有一些原始数据
data = np.array([1, 2, 3, 10, 50, 100])
np.var(data, ddof=0)
mean = np.mean(data)
var = np.std(data, ddof=0)
print('mean', mean)
print('var', var)

# Z得分标准化
scaler_zscore = StandardScaler()
data_zscore = scaler_zscore.fit_transform(data.reshape(-1, 1))

# 画图
plt.figure(figsize=(10, 6))

# 原始数据
plt.subplot(2, 2, 1)
plt.plot(data, label='Original Data')
plt.title('Original Data')
print("Z-Score Normalized Data:\n", data)

# Z得分标准化后的数据
plt.subplot(2, 2, 3)
plt.plot(data_zscore, label='Z-Score Normalized Data', color='green')
plt.title('Z-Score Normalized Data')
print("Z-Score Normalized Data:\n", data_zscore)
print("Z-Score Normalized Data:\n", (data - mean) / var)

plt.tight_layout()
plt.show()
```
